```python
import pandas as pd 
import sqlite3
from datetime import datetime
```

```python
# definir o caminho do arquivo
df = pd.read_json('../../data/data.jsonl', lines=True)
```

```python
# setar o pandas para mostrar todas as colunas
pd.options.display.max_columns = None
```

```python
# adicionar colunas com valor fixo

df['_source'] = "https://lista.mercadolivre.com.br/tenis-corrida-masculino"
```

```python
# adiciona coluna com valor _data_coleta
df['_data_coleta'] = datetime.now()
```

```python
# tratar os valores numericos

# Preços antigos
df['old_price_reais'] = df['old_price_reais'].fillna(0).astype(float)
df['old_price_centavos'] = df['old_price_centavos'].fillna(0).astype(float)

# Preços Novos
df['new_price_reais'] = df['new_price_reais'].fillna(0).astype(float)
df['new_price_centavos'] = df['new_price_centavos'].fillna(0).astype(float)

# Visualizações
df['reviews_rating_number'] = df['reviews_rating_number'].fillna(0).astype(float)

```

```python
# remove parenteses do "reviews_amount"

df['reviews_amount'] = df['reviews_amount'].str.replace(r'[\(\)]', '', regex=True) 
df['reviews_amount'] = df['reviews_amount'].fillna(0).astype(int)
```

```python
# Calcular os valores Totais

# Preços antigos
df['old_price'] = df['old_price_reais'] + df['old_price_centavos'] / 100

# Preços novos 
df['new_price'] = df['new_price_reais'] + df['new_price_centavos'] / 100
```

```python
# remover colunas 
df.drop(columns=['old_price_reais', 'old_price_centavos', 'new_price_reais', 'new_price_centavos'])
```
